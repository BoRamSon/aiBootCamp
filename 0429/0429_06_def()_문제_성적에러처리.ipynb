```python
""" EX:))
def myfunc() :
    if error1 :
        return -1
    if error2 :
        return -2
"""


score_list = []

# ================================================
# ì…ë ¥ + validation + ì´ì™• ê³„ì‚° ê¹Œì§€?!

def getNumber(subject) :
    number = input(f"{subject} : ")
    while isDigit(number) == False :
        print("ìˆ«ìë§Œ ì…ë ¥í•´ì£¼ì„¸ìš”.")
        number = input("ì •ìˆ˜ë¥¼ ì…ë ¥í•˜ì„¸ìš” : ")
    return int(number)

# input ìœ¼ë¡œ ë°›ëŠ” ëª¨ë“  ë°ì´í„°ëŠ” string íƒ€ì…
# ord í•¨ìˆ˜ë¥¼ í†µí•´ì„œ ìˆ«ìì¸ì§€ ì•„ë‹Œì§€ íŒë‹¨ ê°€ëŠ¥
# ë¬¸ìì—´ì„ ë°›ì•„ì„œ í•œê¸€ìì”© "0" ê³¼ "9" ì‚¬ì´ì— ìˆëŠ” ì§€ í™•ì¸
# ê¸€ì ì¤‘ì— í•˜ë‚˜ë¼ë„ 0 - 9 ì‚¬ì´ì— ì¡´ì¬í•˜ì§€ ì•Šìœ¼ë©´ ì—ëŸ¬(ìˆ«ìê°€ ì•„ë‹Œ ê²ƒìœ¼ë¡œ íŒë‹¨)
def isDigit(number) :
    for i in range(0,len(number)) :
        if ord(number[i]) < ord("0") or ord(number[i]) > ord("9") :
            return False

    return True


# ì…ë ¥í•œ ì ìˆ˜ê°€ 0 ~ 100 ì‚¬ì´ì— ìˆëŠ” ìˆ«ìì¸ì§€ í™•ì¸
def getScore(subject = "êµ­ì–´" , limit = 100) :
    n = getNumber(subject)
    while n < 0 or n > limit :
        print(f"0 ~ {limit} ì‚¬ì´ì˜ ê°’ì„ ì…ë ¥í•˜ì„¸ìš”")
        n = getNumber(subject)
    return n

# êµ­ì–´ , ì˜ì–´ , ìˆ˜í•™ ì„±ì  í•©ì‚°
def getSum(s) :
    return s["kor"] + s["eng"] + s["mth"]

# í•©ì‚° ì„±ì ì˜ í‰ê·  ê³„ì‚°
def getAvg(s) :
    return s["sum"] // 3

# ì„±ì  í‰ê· ì— ë”°ë¥¸ ë“±ê¸‰ ì„¤ì •
def getGrade(s) :
    if   s["avg"] >= 80 :
        return "ìˆ˜"
    elif s["avg"] >= 70 :
        return "ìš°"
    elif s["avg"] >= 60 :
        return "ë¯¸"
    elif s["avg"] >= 50 :
        return "ì–‘"
    else :
        return "ê°€"

def append() :
    s = {}
    
    s["name"]  = input("ì´ë¦„ ì…ë ¥ : ")
    s["kor"]   = getScore("êµ­ì–´")
    s["eng"]   = getScore("ì˜ì–´")
    s["mth"]   = getScore("ìˆ˜í•™")
    s["sum"]   = getSum(s)
    s["avg"]   = getAvg(s)
    s["grade"] = getGrade(s)
    
    score_list.append(s)


# ================================================
# ì¶œë ¥

def output() :
    for s in score_list :   # score_list ë¡œë¶€í„° í•˜ë‚˜ì”© s ë¼ëŠ” ë³€ìˆ˜ì— ì „ë‹¬
        print(f"ì´ë¦„ : {s["name"]}",end="\t")
        print(f"êµ­ì–´ ì„±ì  : {s["kor"]}",end="\t")
        print(f"ì˜ì–´ ì„±ì  : {s["eng"]}",end="\t")
        print(f"ìˆ˜í•™ ì„±ì  : {s["mth"]}",end="\t")
        print(f"ì´ ì  : {s["sum"]}",end="\t")
        print(f"í‰ ê·  : {s["avg"]:.0f}",end="\t")
        print(f"ë“± ê¸‰ : {s["grade"]}")


    




def start() :
    while True :
        print("[1] ì¶”ê°€")
        print("[2] ì¶œë ¥")
        print("[0] ì¢…ë£Œ")
        sel = input("ë©”ë‰´ ì„ íƒ : ")
        if sel == "1" :
            append()
        elif sel == "2" :
            output()
        elif sel == "0" :
            print("í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
            return
        else :
            print("ë©”ë‰´ ì…ë ¥ì„ ì˜ëª»í•˜ì…¨ìŠµë‹ˆë‹¤.")
            
start()


```
