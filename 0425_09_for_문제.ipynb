## 🟢 같이하는 예제 - 1~10까지 중복 합계 구하기

- 변수 : 1~10숫자 세는 변수
  - 더해지는 값 : 누적값을 저장할 변수가 필요하다. 누적이 된다. for문 밖에서 초기화 되어야 합니다.
  - 0 + 1
  - 0 + 1 + 2
  - 0 + 1 + 2 + 3
  - 이렇게 반복되는데 반복식을 세우는게 굉장히 중요합니다.




```python
sum_result = 0   # for문 밖에서 초기화 되어야 합니다

for i in range(1,11):
  sum_result += i
  print(f'i = {i} | 중복 더하기 결과 = {sum_result}')

```

```python
sum_result2 = 0   # for문 밖에서 초기화 되어야 합니다

# limit = 10
limit = int(input('제한 수를 입력해주세요'))

for i in range(1,limit+1):
  sum_result2 += i
  print(f'i = {i} | 중복 더하기 결과 = {sum_result2}')

```

## 🟢 문제풀기

- 정수를 5개 입력받아서 합계를 구하라

```python
sum_result4 = 0   # for문 밖에서 초기화 되어야 합니다

for i in range(1,6):
  sum_value = int(input('입력해주세요'))
  sum_result4 += sum_value
  print(f'입력값 = {sum_value} | 중복 더하기 결과 = {sum_result4}')

# 틀림
```
