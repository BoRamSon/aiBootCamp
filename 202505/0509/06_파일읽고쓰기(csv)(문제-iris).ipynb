```python
# Ïñ¥Îñ§ Îç∞Ïù¥ÌÑ∞Î•º Í∞ÄÏßÄÍ≥† ÎÇòÏò§ÎäîÏßÄ ÌôïÏù∏ÌïúÎã§.
# f = open("datas/iris.csv", 'r')
# lines = f.readlines()
# print(lines)
# f.close()


class ReadFileAndPrint:
  def __init__(self):
    self.sepal_length = []
    self.sepal_width = []
    self.petal_length = []
    self.petal_width = []

  def read_file(self):
    f = open("datas/iris.csv", 'r')
    lines = f.readlines() 
    for i in lines:
      seperator = i.split(',')
      self.sepal_length.append(seperator[0])
      self.sepal_width.append(seperator[1])
      self.petal_length.append(seperator[2])
      self.petal_width.append(seperator[3])
    f.close()

    # print(f"Ï∂îÏ∂ú ÏôÑÎ£å sepal_length: {self.sepal_length.pop(0)}")
    # print(f"Ï∂îÏ∂ú ÏôÑÎ£å sepal_length: {self.sepal_length}")
    # print(f"Ï∂îÏ∂ú ÏôÑÎ£å sepal_width: {self.sepal_width}")
    # print(f"Ï∂îÏ∂ú ÏôÑÎ£å petal_length: {self.petal_length}")
    # print(f"Ï∂îÏ∂ú ÏôÑÎ£å petal_width: {self.petal_width}")  

  def calcul_average(self, list):
    name = list[0]
    list.pop(0)  # listÏóê ÏòÅÌñ•ÏùÑ ÎØ∏Ïπ©ÎãàÎã§. Ïã§Ï†ú listÏóê ÏòÅÌñ•ÏùÑ ÎØ∏ÏπòÍ≥† Í∞ÄÏ†∏ÏôÄÏÑú Î∞òÌôòÌï©ÎãàÎã§.
    new_list = []
    for i in list:
      new_list.append(float(i))
    average = sum(new_list) / len(list)
    print(f"\"{name}\"Ïùò ÌèâÍ∑†ÏùÄ {average:.2f}")
    return average    

  def calcul_average_result(self):
    self.calcul_average(self.sepal_length)
    self.calcul_average(self.sepal_width)
    self.calcul_average(self.petal_length)
    self.calcul_average(self.petal_width)

    # print(f"Ï∂îÏ∂ú ÏôÑÎ£å sepal_length: {self.sepal_length}")
    # print(f"Ï∂îÏ∂ú ÏôÑÎ£å sepal_width: {self.sepal_width}")
    # print(f"Ï∂îÏ∂ú ÏôÑÎ£å petal_length: {self.petal_length}")
    # print(f"Ï∂îÏ∂ú ÏôÑÎ£å petal_width: {self.petal_width}")  

  def menu(self):
    print(f"1. ÌååÏùº Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú")
    print(f"2. Ï∂úÎ†•")
    print(f"0. Ï¢ÖÎ£å")

  def start(self):
    print(self.menu())
    while True:
      sel = input('Î©îÎâ¥ Î≤àÌò∏Î•º ÏûÖÎ†•ÌïòÏÑ∏Ïöî. : ')
      if sel == '1':
        self.read_file()      
      if sel == '2':
        self.calcul_average_result()
      if sel == '0':
        return print(' üëã Ï†ïÏÉÅÏ†ÅÏúºÎ°ú Ï¢ÖÎ£åÎêòÏóàÏäµÎãàÎã§.')

r = ReadFileAndPrint()
r.start()



```
