## ğŸŸ¢ ì œëŒ€ë¡œ í•´ë³´ê¸°

```python
person = {'ì´ë¦„': 'í™ê¸¸ë™', 'ë‚˜ì´': 20, 'ì„±ë³„': 'ë‚¨ì', 'ì·¨ë¯¸': ['ë…ì„œ', 'ì˜í™” ê°ìƒ'], 'ì§ì—…': 'í•™ìƒ'}

# items() ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ë”•ì…”ë„ˆë¦¬ì˜ ëª¨ë“  í‚¤-ê°’ ìŒì„ ì¶œë ¥
for key, value in person.items():
    print(f"{key}: {value}")


# values() ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ë”•ì…”ë„ˆë¦¬ì˜ ëª¨ë“  ê°’ë§Œ ì¶œë ¥
for value in person.values():
    print(value)


print(person['ì´ë¦„'])  # í™ê¸¸ë™
print(person['ë‚˜ì´'])  # 20
print(person['ì„±ë³„'])  # ë‚¨ì
```

```python
#### ğŸŸ¡ change 

person['ë‚˜ì´']=30 # ë‚˜ì´ë¥¼ 30ìœ¼ë¡œ ë³€ê²½
print(person)  
```

```python
#### ğŸŸ¡ update

person.update({'ì´ë¦„': 'ê¹€ì² ìˆ˜', 'ì„±ë³„': 'ì—¬ì'}) # ì´ë¦„ê³¼ ì„±ë³„ì„ ë³€ê²½
print(person)  # {'ì´ë¦„': 'ê¹€ì² ìˆ˜', 'ë‚˜ì´': 30, 'ì„±ë³„': 'ì—¬ì', 'ì·¨ë¯¸': ['ë…ì„œ', 'ì˜í™” ê°ìƒ'], 'ì§ì—…': 'í•™ìƒ'}
```

```python
#### ğŸŸ¡ fromkeys() ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ìƒˆë¡œìš´ ë”•ì…”ë„ˆë¦¬ ìƒì„±

# keys: ë”•ì…”ë„ˆë¦¬ì˜ í‚¤ ëª©ë¡
# value: ëª¨ë“  í‚¤ì— ê³µí†µì ìœ¼ë¡œ í• ë‹¹ë  ê°’ â† ì¤‘ìš”!

keys = ['ì´ë¦„', 'ë‚˜ì´', 'ì„±ë³„']
values = ['í™ê¸¸ë™', 20, 'ë‚¨ì']
new_person = dict.fromkeys(keys, values)
print(new_person)  # {'ì´ë¦„': ['í™ê¸¸ë™', 20, 'ë‚¨ì'], 'ë‚˜ì´': ['í™ê¸¸ë™', 20, 'ë‚¨ì'], 'ì„±ë³„': ['í™ê¸¸ë™', 20, 'ë‚¨ì']}
```

```python
# zip() í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ë‘ ê°œì˜ ë¦¬ìŠ¤íŠ¸ë¥¼ ë¬¶ì–´ ë”•ì…”ë„ˆë¦¬ ìƒì„±

keys2 = ['ì´ë¦„', 'ë‚˜ì´', 'ì„±ë³„']
values2 = ['í™ê¸¸ë™', 20, 'ë‚¨ì']
new_person2 = dict(zip(keys2, values2))
print(new_person2)  # {'ì´ë¦„': 'í™ê¸¸ë™', 'ë‚˜ì´': 20, 'ì„±ë³„': 'ë‚¨ì'}
```
