### ğŸŸ¡ ë°°ìš°ê¸°

```python
# ì—¬ê¸°ì—ì„œëŠ” centerë¥¼ ì¼ìŠµë‹ˆë‹¤. spaceì— ëŒ€í•´ì„œ 

n = 4  # ê°€ìš´ë° ì¤„ì˜ ë³„ ê°œìˆ˜ë¥¼ ì •í•˜ëŠ” ìˆ«ì (4 -> ê°€ìš´ë° 7ê°œ)

# ìœ„ìª½ í”¼ë¼ë¯¸ë“œ (1, 3, 5, 7ê°œ)
for i in range(n):
    stars = '*' * (2 * i + 1)
    print(stars.center(2 * n - 1))

# ì•„ë˜ìª½ í”¼ë¼ë¯¸ë“œ (5, 3, 1ê°œ)
for i in range(n - 2, -1, -1):
    stars = '*' * (2 * i + 1)
    print(stars.center(2 * n - 1))
```
