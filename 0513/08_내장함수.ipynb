# 🟩 내장함수

- 보통 자주 사용하는 내장함수들에 대해서 알아보자!


## 🟢 자주쓰는 내장함수 목록 요약정리

| 함수            | 설명                 |
| ------------- | ------------------ |
| `abs()`       | 절댓값 반환             |
| `all()`       | 모두 참이면 True        |
| `any()`       | 하나라도 참이면 True      |
| `ord()`       | 문자 → 유니코드 정수 변환    |
| `chr()`       | 유니코드 정수 → 문자 변환    |
| `dir()`       | 객체의 모든 속성과 메서드 리스트 |
| `divmod()`    | 몫과 나머지를 튜플로 반환     |
| `enumerate()` | 인덱스와 값을 함께 반환      |
| `eval()`      | 문자열을 코드로 실행        |
| `filter()`    | 조건에 맞는 요소만 걸러냄     |
| `max()`       | 최대값 반환             |
| `min()`       | 최소값 반환             |
| `pow()`       | 거듭제곱 계산            |
| `round()`     | 반올림                |


```python
# -------------------------------
# 절댓값 (absolute value)
print('< abs >')  
print(abs(-4))    # 4
print(abs(10))    # 10


# -------------------------------
# all() : 모든 요소가 참이면 True
print('< all >') 
print(all([1, 2, 3]))       # True
print(all([1, 2, 3, 0]))    # False (0은 False 취급)
print(all(['a', 'b', 'c'])) # True
print(all(['a', 'b', '']))  # False (''는 False)

# all()은 빈 리스트일 경우 True를 반환합니다
print(all([]))              # True


# -------------------------------
# any() : 하나라도 참이면 True
print('< any >')
print(any([1, 2, 3]))       # True
print(any([1, 2, 3, 0]))    # True
print(any([0, "", False]))  # False
print(any([]))              # False


# -------------------------------
# ord() / chr()
print('< ord / chr >')
# ord : 문자를 유니코드 정수로
print(ord('A'))  # 65
print(ord('a'))  # 97

# chr : 유니코드 정수를 문자로
print(chr(65))   # 'A'
print(chr(97))   # 'a'


# -------------------------------
# dir() : 객체의 속성 및 메서드 목록을 반환
print('< dir >')
print(dir([1, 2, 3]))   # 리스트가 가진 메서드들
print(dir(dict()))      # 딕셔너리가 가진 메서드들


# -------------------------------
# divmod(a, b) : 몫과 나머지를 튜플로 반환
print('< divmod >')
print(divmod(5, 2))   # (2, 1) -> 5를 2로 나눈 몫과 나머지
print(divmod(9, 3))   # (3, 0)


# -------------------------------
# enumerate() : 인덱스와 값을 함께 반환하는 반복문
print('< enumerate >')
for i, c in enumerate("Life is egg"):
    print(i, c)


# -------------------------------
# eval() : 문자열을 파이썬 코드로 실행
print('< eval >')
result = eval('1 + 10 + 3')   # 14
print(result)

# 변수도 쓸 수 있음
x = 5
print(eval('x + 10'))         # 15


# -------------------------------
# filter(함수, iterable) : 조건을 만족하는 요소만 필터링
print('< filter >')
a = [3.4, -1, 2, 9, 8, 7, 12, 15, 21]

# 일반 함수 사용
def is_positive(x):
    return x > 0

po_list = list(filter(is_positive, a))
print(po_list)  # [3.4, 2, 9, 8, 7, 12, 15, 21]

# lambda (익명 함수) 사용
po_list2 = list(filter(lambda x: x > 0, a))
print(po_list2)


# -------------------------------
# max() / min() : 최대값 / 최소값
print('< max / min >')
print(f"최대값: {max(a)} | 최소값: {min(a)}")


# -------------------------------
# pow(x, y) : x의 y제곱을 계산 (x ** y와 동일)
print('< pow >')
print(pow(2, 4))  # 2^4 = 16
print(pow(3, 3))  # 3^3 = 27


# -------------------------------
# round() : 반올림 (소수점 자리수 지정 가능)
print('< round >')
print(round(3.14159))      # 3
print(round(3.14159, 2))   # 3.14
print(round(1.5555, 3))    # 1.556

# 참고: 파이썬의 round는 "은행 반올림" 방식 (0.5는 짝수로 반올림)
print(round(0.5))          # 0
print(round(1.5))          # 2
```
