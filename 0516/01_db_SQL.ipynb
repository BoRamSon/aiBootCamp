## 🟩 테이블 구조를 확인하는 명령어

#### 🟡 desc dept; (desc 테이블명)
- 테이블 구조 상세파악 (from 'desc dept;')
  
    ```markdown
      +--------+-------------+------+-----+---------+-------+
      | Field  | Type        | Null | Key | Default | Extra |
      +--------+-------------+------+-----+---------+-------+
      | deptno | int         | NO   | PRI | NULL    |       |
      | dname  | varchar(14) | YES  |     | NULL    |       |
      | loc    | varchar(13) | YES  |     | NULL    |       |
      +--------+-------------+------+-----+---------+-------+
    ```
  - Field : 열에 명칭이 들어간다.

  - Type : 데이터 타입이 존재한다. small int int bigint,,,,
  
  - char
    - 고정길이 문자열, 1000 byte 이내
    - gender char (10)  gender = 'Y'
    - y ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ 10개의 메모리를 다 차지
    - where gender = 'y'
    - where gender = 'y' 다른 dbms마다 다른데 대부분은 데이터에서 한해서 대소문자 구분을 하는데 mysql은 못한다.
      - 오라클의 경우에는 where gender = 'y        '
      - 또는 where trip(gender) = 'y'
    - 데이터 길이가 정해져 있을 때 사번, 학번, 성별, 연도
  
  - varchar (Variable Charact '바차'라고 발음)
    - 가변길이 문자열, 용량을 지정해도 실제 데이터만큼만 사용을 한다.
    - varchar(100) 'y'가 들어ㅇ가면 실제 데이터길이 1만 차지한다.
    - 2000byte 까지 앞부분에 데이터길이를 별도로 저장한다. 그래서 상황에 맞춰서 잘 선택하고,
    - 사용자 아이디, 게시글 타이틀 정도할 때 유용 (너무 긴 내용 못씀!)

  - text, long text 
    - 2GB 
    - 게시판 게시글 같은 곳에 사용

  - Null : 널 허용 여부
  - Key : pri - primary key : 중복불가, null불가 조건을 만족해야한다.
  - Default : 특별한 값을 입력하지 않았을 떄 기본값으로 저장된다.

<br><br><br>

```markdown
+--------+-------------+------+-----+---------+-------+
| Field  | Type        | Null | Key | Default | Extra |
+--------+-------------+------+-----+---------+-------+
| deptno | int         | NO   | PRI | NULL    |       |
| dname  | varchar(14) | YES  |     | NULL    |       |
| loc    | varchar(13) | YES  |     | NULL    |       |
+--------+-------------+------+-----+---------+-------+
```


| 항목          | 설명                                                                                                                                                                                                                                                                                |
| ----------- | --------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Field**   | \*\*컬럼 이름(필드명)\*\*입니다. 테이블에 존재하는 각 열(데이터 항목)의 이름이에요.<br>예: `deptno`, `dname`, `loc` 등.                                                                                                                                                                                            |
| **Type**    | **데이터 타입과 크기**를 나타냅니다.<br>- `int`: 정수<br>- `varchar(14)`: 최대 14글자까지 저장 가능한 문자열<br>- `date`, `datetime`, `float`, `decimal`, `text`, `char(n)` 등 여러 타입이 있습니다.<br>**varchar**는 글자 수가 달라도 저장 공간을 유동적으로 사용해서 효율적입니다.                                                                  |
| **Null**    | 이 필드에 \*\*값이 없을 수 있는지(NULL 가능 여부)\*\*를 나타냅니다.<br>- `YES`: 비워도 됨 (NULL 저장 가능)<br>- `NO`: 반드시 값이 있어야 함 (NOT NULL)<br><br>📌 NULL은 “빈칸”과는 다릅니다. “정해지지 않음”이라는 의미예요.                                                                                                                   |
| **Key**     | **이 컬럼이 어떤 종류의 키로 사용되는지** 알려줍니다.<br>- `PRI` (Primary Key): 기본키, 테이블에서 유일하고 NULL이 될 수 없음. 중복 불가.<br>- `UNI` (Unique): 중복은 허용되지 않지만 NULL은 허용될 수 있음.<br>- `MUL` (Multiple): 중복 가능하며, 인덱스가 생성됨. 외래키(foreign key)나 검색 속도 향상을 위한 키일 수 있어요.<br><br>🔑 이건 데이터의 **식별자 또는 검색 속도**와 관련 있습니다. |
| **Default** | **기본값**이에요. 이 컬럼에 값을 넣지 않으면 자동으로 저장되는 값입니다.<br>예: `status` 컬럼이 있고 기본값이 `'active'`라면 값을 입력 안 해도 `'active'`가 저장됩니다.<br>값이 `NULL`이면 기본값이 따로 설정되어 있지 않다는 뜻입니다.                                                                                                                        |
| **Extra**   | **추가 속성**을 의미합니다.<br>주로 다음과 같은 특별한 설정이 표시됩니다:<br>- `auto_increment`: 값을 자동으로 1씩 증가시킴 (주로 Primary Key에서 사용)<br>- `on update CURRENT_TIMESTAMP`: 값이 자동으로 갱신됨<br>없으면 비어 있음.                                                                                                          |

#### 🟡 MySQL 또는 일반적인 관계형 데이터베이스(RDBMS)에서 SQL 명령어는 기능에 따라 세 가지 주요 범주

| 분류  | 이름         | 대표 명령어                                 | 설명                 |
| --- | ---------- | -------------------------------------- | ------------------ |
| DDL (Data Definition Language) | 데이터 정의 언어  | `CREATE`, `ALTER`, `DROP`, `TRUNCATE`  | 테이블 구조 정의/변경/삭제    |
| DML (Data Manipulation Language) | 데이터 조작 언어  | `SELECT`, `INSERT`, `UPDATE`, `DELETE` | 데이터 조회, 삽입, 수정, 삭제 |
| DCL (Data Control Language) | 데이터 제어 언어  | `GRANT`, `REVOKE`                      | 사용자 권한 설정 및 회수     |
| TCL (ransaction Control Language) | 트랜잭션 제어 언어 | `COMMIT`, `ROLLBACK`, `SAVEPOINT`      | 데이터 변경 확정 또는 취소    |

<br>

#### ⚫️ DDL (Data Definition Language) - 데이터 정의 언어
  - → 테이블 구조를 생성·변경·삭제할 때 사용하는 명령어

| 명령어        | 설명                                         |
| ---------- | ------------------------------------------ |
| `CREATE`   | 테이블, 데이터베이스, 뷰 등을 **생성**                   |
| `ALTER`    | 테이블 구조를 **수정** (컬럼 추가, 타입 변경 등)            |
| `DROP`     | 테이블이나 데이터베이스를 **삭제**                       |
| `TRUNCATE` | 테이블의 모든 데이터를 **삭제**하면서 **초기화** (로그 남기지 않음) |
| `RENAME`   | 테이블의 이름을 **변경**                            |

<br>

#### ⚫️ DML (Data Manipulation Language) - 데이터 조작 언어
  - → 테이블에 들어있는 데이터 자체를 조작할 때 사용

| 명령어      | 설명             |
| -------- | -------------- |
| `SELECT` | 데이터 조회 (읽기)    |
| `INSERT` | 데이터 **삽입**     |
| `UPDATE` | 기존 데이터를 **수정** |
| `DELETE` | 데이터를 **삭제**    |

<br>

#### ⚫️ DCL (Data Control Language) - 데이터 제어 언어
  - → 사용자 권한 관리 및 보안 제어 목적

| 명령어      | 설명                   |
| -------- | -------------------- |
| `GRANT`  | 사용자에게 권한을 **부여**     |
| `REVOKE` | 사용자에게 부여된 권한을 **회수** |

<br>

#### ⚫️ 보너스: TCL (Transaction Control Language) - 트랜잭션 제어 언어

| 명령어         | 설명                   |
| ----------- | -------------------- |
| `COMMIT`    | 변경 내용을 **확정 저장**     |
| `ROLLBACK`  | 변경 내용을 **취소, 되돌림**   |
| `SAVEPOINT` | 롤백할 위치 지정 (부분 롤백 지원) |

## 🟢 emp_def_sample.sql 학습용 DB 데이터를 통해서 학습하기


````python
-- emp 테이블의 내용을 다 보여줌 - 14건
select * from emp;

-- 데이터 전체 몇건인가??? 
select count(*) from emp; 

-- * : 아스테리스크 모든 필드
select empno, enamem, job from emp

--  aliasing - 별명지정하여 테이블을 수정해서 부를 수 있다.
select empno, enamem, job 
from emp as e;

-- 이렇게 aliasing을 통해서 만들면 빨라집니다!!!
-- information_schema 디비에 사용자가 만드는 테이블 정보가 저장된다. 
-- aliasing 안쓰면 저 디비 다 검색해서 정보를 읽어오고, aliasing을 하면 임시로 후다닥 읽어옵니다(cashing)
-- 디비정보를 메리에 붙러놓고 작업 속도가 유리하다
select e.empno, e.enamem, e.job 
from emp as e;

-- as 생략 가능 / count(필드명) 대부분의 dbms는 null값이 있을 경우 카운트 안함
-- 있을 경우, 카운트 안함
-- count(*) - 필드중에 null 값이 아예 없는 필드를 기준으로 제일 많은 데이터 카운트를 가져와라
select count(e.comm) from emp e;

/*
	< null 값 >
    알 수 없음, 빈값이 아님. 
    파이썬의 경우는 None. 
    수학적으로 무한대의 의미를 갖는다. 
    null + 1  =>  null (무한대)
    null + 1  =>  null (무한대)
    
	(수학연산 다 가능 (+, -0, *, /, sin, cosin, tan, round .........)
    
    null값에 연산을 하면 결과는 null
    파이썬의 if문, 함수를 써서 처리가능
    DBMS의 쿼리는 AWSI 표준이 있어서 비슷한데 함수는 각자 다르다
	
    IFNULL(필드명, 기본값) 만일 필드명에서 값이 NULL이 아닌 값이 오면 그 값을 주고 NULL이면 기본값을 부여한다.	
*/

-- select empno, sal+comm from emp; 대문자로 해봅시다
-- SELECT EMPNO, sal, comm, SAL+COMM FROM EMP;


-- 연산을 통해서 새로운 컬럼을 만들었음, 수식이 필드명으로 나옴
-- 필드명도 aliasing을 통해서 다시 부여할 수 있음. 
select empno, sal, comm, sal+IFNULL(comm, 0) as totla_sal
from emp;
````

## 🟢 SQL 이론

- 무결성: 결함이 없는 성격
- 중복성 배제: 데이터의 중복 배제
- 일관성: 데이타가 중복도 없고 일관성이 있다.

- primary key 개념
  - 특정 필드 또는 어러개의 필드를 묶어서 중복성을 배제하고 
  - null값도 갖지 못하도록 제약하는 특성을 갖는다.

- 사번, 학번, 주민번호 등 (주민번호는 이미 중복된 사람 20000명 있음)은 사용 못함.
- 필드의 성격에 auto_increment 자동증가, mssql에서는 일련번호, 오라클은 시쿼스라는 객체.

- mysql은 auto_increment 속성이 있는 필드가 primary key여야 한다.


```sql
use mydb;

desc emp;

-- empno 필드가 primary key로 되어있다.
insert into emp (empno, ename) values (8000, '홍길동');
insert into emp (empno, enaee) values (8000, '임꺽정');
-- 8000번이 이미 있어서 에러발생
-- Error Code: 1062. Duplicate entry '8000' for key 'emp.PRIMARY'

-- 테이블을 만들 때 
```


- 테이블을 만들 때, primary key를 지정할 수도 있고, 테이블 다 만들고 지정할 수도 있다.
- 일부 데이터를 먼저 테이블에 넣어놓고 primary key 지정할 때
- 이미 들어가 있는 데이터가 중복이되거나 nill 값이 있으면 primary key를 지정 못함.
- 필드 하나로 primary 키를 지정할 수도 있고 필요하면 여러 개의 필드를 묶어서 하나의 primary key를 지정할 수 있다.
- 단, 하나의 테이블에 한개만 지정 가능하다.

- 기본적으로 검색이 순차검색이기 때문에 한개씩 읽어서 검색한다.
- oracle은 병행처리, 한번에 여러개 읽는다.
- 색인순차검색, 색인표를 만들어서 검색한다. PRIMARY KEY를 지정하면 자동으로 색인(INDEX)를 만들어준다.
