```python
worker_list = [
                {"name" : "í™ê¸¸ë™" , "work_time" : 30 , "per_pay" : 20000} ,              #ê³µìš©ë³€ìˆ˜ë¡œ ì¡´ì¬ , í…ŒìŠ¤íŠ¸ìš© ë°ì´í„° ì„  ì…ë ¥
                {"name" : "ê³ ê¸¸ë™" , "work_time" : 20 , "per_pay" : 30000} ,
                {"name" : "ê¹€ê¸¸ë™" , "work_time" : 50 , "per_pay" : 20000}
               ]



def append() : #ë°ì´í„° ì¶”ê°€ í•¨ìˆ˜ ì‘ì„±
    worker = {}
    worker["name"]      =     input("ì´ë¦„ : ")
    worker["work_time"] = int(input("ê·¼ë¬´ ì‹œê°„ : "))
    worker["per_pay"]   = int(input("ì‹œê°„ ë‹¹ ê¸‰ì—¬ì•¡ : "))
    worker["pay"]       = 0     # ì£¼ê¸‰ ê³„ì‚°ì„ ìœ„í•´ì„œ ì…ë ¥
    
    worker_list.append(worker)  # ëª©ë¡ì— ì¶”ê°€í•˜ê¸°

def process(worker) :       # ì£¼ê¸‰ ê³„ì‚° í•¨ìˆ˜ (dic íƒ€ì… ë°ì´í„°ë¥¼ ê°€ì ¸ì™€ì„œ ë°˜í™˜í•˜ëŠ” ë°©ë²•) , ë§¤ê°œë³€ìˆ˜ë¡œ ê°’ì„ ë°›ì•„ì˜¤ë©´ ì™¸ë¶€ë¡œ ì „ë‹¬ ì•ˆë˜ëŠ” ê²ƒì´ ì›ì¹™
    worker["pay"] = worker["work_time"] * worker["per_pay"]
    
def process_main() :        # process í•¨ìˆ˜ëŠ” í•œ ëª…ë¶„ì— ëŒ€í•œ ì²˜ë¦¬ í•¨ìˆ˜ì´ê¸°ì— ë³„ë„ë¡œ í•¨ìˆ˜ë¥¼ ë§Œë“¤ì–´ ë¦¬ìŠ¤íŠ¸ ë‚´ì˜ ì „ì²´ ê³„ì‚°ì„ ìœ„í•œ í•¨ìˆ˜ë¥¼ ë³„ë„ ì‘ì„±
    for w in worker_list :
        process(w)

def output() :
    for w in worker_list :
        print(f"{w["name"]}" , end="\t")
        print(f"{w["work_time"]}" , end="\t")
        print(f"{w["per_pay"]}" , end="\t")
        print(f"{w["pay"]}" , end="\t")
        print() # ì¤„ë°”ê¿ˆ ì½”ë“œ

def main_menu() :
    while(True) :              # while(True) : ë¬´í•œ ë£¨í”„ ì½”ë“œ - ì¢…ë£Œë¥¼ í•˜ì§€ ì•ŠëŠ”ë‹¤. ì¢…ë£Œë¥¼ ì›í•˜ë©´ break ë¬¸ì„ ì“°ê±°ë‚˜ return êµ¬ë¬¸ì„ ì‚¬ìš©
        print("[1] ì¶”ê°€")       # return - í•¨ìˆ˜ë¥¼ ì¢…ë£Œí•˜ë©´ì„œ í•¨ìˆ˜ì˜ ì‘ì—… ë‚´ìš©ì„ í•¨ìˆ˜ ì™¸ë¶€ë¡œ ì „ë‹¬í•¨
        print("[2] ì¶œë ¥")       # return ì„ ì“°ê³  ê°’ì„ ì•ˆì£¼ë©´ ê·¸ëƒ¥ í•¨ìˆ˜ê°€ ì¢…ë£Œë¨(í•¨ìˆ˜ ì¢…ë£Œë¥¼ ì˜ë¯¸í•¨)
        print("[3] ê³„ì‚°")
        print("[0] ì¢…ë£Œ")               
        sel = input("ì„ íƒ : ")
        if sel   == "1" :       # ì‚¬ìš©ìê°€ 1 ì„ ì…ë ¥í–ˆì„ ë•Œ
            append()            # ë°ì´í„° ì¶”ê°€ í•¨ìˆ˜ í˜¸ì¶œ
        elif sel == "2" :       # ì‚¬ìš©ìê°€ 2 ë¥¼ ì…ë ¥í–ˆì„ ë•Œ
            output()            # ì¶œë ¥ í•¨ìˆ˜ í˜¸ì¶œ
        elif sel == "3" :       # ì‚¬ìš©ìê°€ 3 ì„ ì…ë ¥í–ˆì„ ë•Œ
            process_main()      # ê³„ì‚° í•¨ìˆ˜ í˜¸ì¶œ
        elif sel == "0" :       # ì‚¬ìš©ìê°€ 0 ì„ ì…ë ¥í–ˆì„ ë•Œ
            print("í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
            return              # í•¨ìˆ˜ ìì²´ë¥¼ ì¢…ë£Œ
        else :
            print("ì˜ëª» ì„ íƒí•˜ì…¨ìŠµë‹ˆë‹¤.")   # ì…ë ¥ ì˜¤ë¥˜ ì²´í¬


main_menu()
```
