# 🟩 Git
  - 버전 관리 시스템
  - 쉽게 말해, 파일(특히 소스코드)의 변경 이력을 저장하고 관리하는 도구입니다. 개발자가 코드를 작성할 때, 언제 어떤 내용이 바뀌었는지 추적할 수 있고, 필요하면 예전 상태로 되돌릴 수도 있습니다.

    - 백업(Backup)

  - 개발자들은 Git을 통해 수많은 소스 코드를 효율적으로 관리


## 🟢 Command Line Interface (CLI)
- 터미널에 명령어를 직접 입력하여 Git을 사용하는 방식
- Git 명령어와 기본적인 리눅스 명령어에 대한 이해가 필요함
- 많은 개발자들이 이 방식을 선호하고 실제로 많이 사용함

- 환경 설정
  - git config --global user.name "boramson"
  - git config --global user.email "boram@example.com"
  - 확인하는 명령어
    - git config --global user.name
    - git config --global user.email

- 기본적인 리눅스 명령어
  - pwd : 현재 디렉토리 위치 확인
  - ls : 폴더 안의 파일/디렉토리 목록 보기
    - (한칸 띄어쓰기 후 아래 추가 명령어를 확인 가능)
    - -a : 숨김 파일과 디렉터리를 함께 표시
    - -l : 파일이나 디렉터리의 상세 정보를 함께 표시
    - -r : 파일의 정렬 순서를 거꾸로 표시
    - -t : 최근 수정된 파일부터 보여줌 (시간순 정렬)
  - cd 명령어 ()
    - cd my_folder   : 해당 폴더경로로 들어가기
    - cd ..   : 뒤로 가기
      - ~  : 
      - .  : 
      - .. : 
  - mkdir 폴더 만들기 (make directory)
    - mkdir (생성할_디렉터리_이름)
  - rm 삭제하기
    - rm -r (삭제할_디렉터리_이름)

## 🟢 터미널 안에서 메모장 쓰는 방법 (Vim)
  - 터미널에서 사용할 수 있는 대표적인 문서 편집기
  - 작업할 디렉토리에 들어간다.
  - 명령어 = vim test.txt (파일이 없으면 생성, 있으면 열기)
  - 터미널에 물결표시의 화면(ex모드)이 열립니다.
  - A(add) 또는 I(insert)키를 눌러서 입력모드 전환!
  - 원하는 내용 입력하기
  - 원하는 내용 입력이 끝난 후 'ESC'키를 통해 나옵니다. (ex모드로 돌아온다.)
  - 터미널 맨 아래에 
    - :w (write)
    - :q (quit) 
    - :wq (write quit) 
    - :q! (강제종료 / 저장하지 않음)
    - 위 명령어를 알맞게 입력 후 enter! 끝!
  - 작성한 txt 단순히 읽어들이기
    - cat test.txt
    - 내용을 확인할 수 있다.

## 🟢 나의 Local에 Git 저장소 만들기
- Git 전용 폴더 만들기
  - mkdir hello-git  (단순 폴더 만들기)
  - cd hello-git  (만든 폴더로 이동)
  - 해당 폴더를 Git 전용 저장소로 만들기
    - 🔥 git init 
    - Initialized empty Git repository in /Users/본인폴더경로/hello-git/.git/
    - .git이라는 숨김폴더가 만들어집니다. 여기에 모든 버전의 정보들이 저장될 것입니다.

- 개념
  - working tree (작업트리)
    - 파일 수정, 저장 등의 작업을 하는 디렉토리
    - 우리 눈에 보이는 디렉터리를 말함
  - stage (스테이지)
    - 버전으로 만들 파일이 대기하는 곳
    - 눈에 보이지 않음
  - repository (저장소)
    - 스테이지에서 대기하고 있던 파일들을 버전으로 만들어 저장하는 장소
    - 눈에 보이지 않음

## 🟢 버전 만들기
  - add : 작업트리에서 스테이지로 추가합니다.
  - commit : 스테이지에서 저장소로 저장 (version 1 만들어짐)

- 현재 나의 깃전용저장소 확인하기
  - (이 깃전용저장소에 test.txt를 넣어놓아보았습니다.)
  - git status
  ```markdown
    On branch main

    No commits yet

    Untracked files:
      (use "git add <file>..." to include in what will be committed)
    	test.txt

    nothing added to commit but untracked files present (use "git add" to track)
  ```

- 버전 올리기
  - add
    - git add .    (수정한 전체 파일 stage에 올리기)
    - git add test.txt  (특정 파일만 stage에 올리기)
  - commit
    - (현재 stage에 있는 모든 파일에 대한 commit이 진행된다.)
    - git commit -m "feat: create test.txt"
      - (-m 메세지 / " " 내부에 메세지 지정)

- 커밋한 버전 확인하기
  - git log 
    ```markdown
      Author: username <useremail@gemail.com>
      Date:   Thu May 22 12:30:30 2024 +0900

      create test.txt   (이게 깃커밋 메세지 입니다.)
    ```
  - git log --stat
    - (파일까지 확인할 수 있습니다.)
    ```markdown
      Author: BoRamSon <onbrson1@gmail.com>
      Date:   Thu May 22 13:05:38 2025 +0900

          create bye.txt

       bye.txt     | 0
       coding.html | 0
       style.css   | 0
       system.txt  | 0
       4 files changed, 0 insertions(+), 0 deletions(-)
    ```
  
## 🟢 직접 해보기
  - 수정 > 저장 > add > commit > log 등 버전내용 확인 과정을 거쳐 봅시다.
    - vim test.txt

## 🟢 변경 사항 확인 방법
  - (add 상태, 즉 stage에 있는 상태의 파일과 최신 버전(이미 올라간 파일)의 내용을 비교할 때)

## 🟢 Git에서 관리 제외 시키기 (.gitignore)
  - 나의 프로젝트 폴더 내 .gitignore 파일을 만듭니다.
  - Git에 넣지 않은 파일이나 폴더의 명칭을 .gitignore 내부에 적어놓으면 Git에서 자동으로 인식을 하지 않습니다.
  - <실습>
    - 각종 파일 만들어보기 
      - touch coding.html style.css system.txt
    - git status
      - new file:   coding.html
	    - new file:   style.css
	    - new file:   system.txt

## 🟢 수정 또는 파일 바로 이전으로 되돌리기
  - git restore 파일명 / git restore --staged 파일명 / git reset HEAD^
  - 1. 이전의 저장했던 내용으로 되돌리고 싶어요.
    - vim bye.txt  (내용 수정)
    - git status (수정된 파일에 대해서 확인)
    - git restore bye.txt
  - 2. stage에 올라간 파일을 취소하고 싶어요.
    - vim bye.txt  (내용 수정)
    - git status (수정된 파일에 대해서 확인)
    - git add bye.txt
    - git status (현재 상태 확인)
    - git restore --staged bye.txt
  - 3. commit한 최신 버전을 취소시켜버리는 방법
    - vim bye.txt  (내용 수정)
    - git commit -am "reset test" (add와 commit을 한번에)
    - git log
    - git reset HEAD^  

## 🟢 reset - 특정 커밋(버전)으로 되돌리기
  - git reset --옵션 `되돌아갈_커밋해시`
    - `--option`
    - `--hard` : 특정 버전의 커밋으로 되돌린 다음 그 이후 버전을 삭제
    - `--soft` : 커밋 이력은 삭제되지만 변경 내용은 스테이지 상태로 남아있음
    - `--mixed` : 커밋 이력은 삭제되지만 변경 내용은 작업트리에 남아있음

  - option --hard
    - git log
    - 원하는 해시 복사하기 commit 76aea9892173126ed7b9d6dwqd2e233d1w52012d53a4ee
    - git reset --hard 76aea9892173126ed7b9d6dwqd2e233d1w52012d53a4ee
    - HEAD is now at ~~~~~~   이런식으로 왔다는 메세지가 뜨게됩니다.


## 🟢 revert - 특정 커밋(버전)으로 되돌리기
  - 기존 커밋을 삭제하지 않고, 특정 커밋으로 되돌리는 새로운 커밋을 추가하는 방식
  - 커밋 기록을 변경하지 않고, 안전하게 작업을 취소하는 방법
  - 협업 환경에서 주로 사용 (삭제가 없어서 안정하기 때문)
  - 하지만 로그가 지저분해진다는 단점이 있습니다.
  - (생길 수 있는 충돌에 대해서 대비할 필요가 있습니다. 어떠한 충돌이 있을 수 있을까?)
  - git log   (돌아갈 버전 파악)
  - git revert  (해시값)
  - git revert 5b5caeww6Ewqr9d63b7321229e412314
  - 자동으로 원하던 버전을 최신으로 가져와서 버전을 별도로 만듭니다.



## 🟢 Branch

  - git merge B

# 🟩 Github


## 🟢 Github와 함께 활용하기

1. local 저장소에 나의 프로젝트 폴더(ex. hello-git) 생성
2. 해당 프로젝트 폴더 내 test 파일 만들고 수정하고 add, commit
   - git init
   - vim cnn.txt (A라는 내용 입력)
   - git add cnn.txt
   - git commit -m "add A"
3. 원격 저장소 생성 (Github 저장소 생성)
   - repository name : whatisgit
   - create repository
4. Github 원격 저장소와 연결
   - git remote add origin <복사한 주소>
      - remote: 원격 저장소를 의미
      - add: 추가해
      - origin 이라는 변수를 추가해라! (다른 이름도 무관하다)
      - <복사한 주소>: 나의 Github 원격 Repository 주소
   - 만약 저장소가 바뀌어서 저장소를 바꿔야 할 경우 origin을 비워주고 다시 시도해야함.
     - git remote remove origin   
5. git push (github로 올림)
   - git push -u origin main
     - -u : 지역저장소의 브랜치를 원격저장소에 연결
     - origin : 원격저장소 주소
     - main : 원격저장소의 브랜치 이름
     - (한 번 이상 원격저장소에 업로드한 경우, 다음 명령어로 push가능)
6. Practice
   - vim conn.txt
   - git add conn.txt
   - git commit -m "add b"
   - git push
7. 원격저장소 내 파일 나의 로컬저장소로 가져오기
   - (원격저장소에 있는 파일을 다른 사용자가 수정하여 나의 로컬저장소의 파일과 차이가 있는 경우)
   - Pull
   - git pull origin main
     - 결과: create mode 100022 conn2.txt
   - `ls` 로 확인하기
