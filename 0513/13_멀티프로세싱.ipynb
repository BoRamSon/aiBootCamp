# 🟩 멀티프로세싱 이론

| 항목       | 멀티프로세싱 (`multiprocessing`)   | 멀티스레딩 (`multithreading`)        |
| -------- | ---------------------------- | ------------------------------- |
| ✅ 정의     | 여러 **프로세스**를 동시에 실행          | 하나의 프로세스 내에서 여러 **스레드**를 동시에 실행 |
| ✅ 메모리    | 각 프로세스는 **독립된 메모리 공간** 사용    | 스레드는 **공유 메모리** 사용              |
| ✅ CPU 활용 | 진짜 병렬 처리 (멀티코어 CPU 활용)       | GIL 때문에 병렬 처리 어려움 (Python에서만)   |
| ✅ 안정성    | 하나의 프로세스가 죽어도 다른 프로세스는 영향 없음 | 하나의 스레드가 죽으면 전체 프로세스에 영향 가능성    |
| ✅ 성능     | CPU 계산 중심 작업에 강함             | I/O 중심 작업 (파일, 네트워크 등)에 강함      |
| ✅ 예시     | 이미지 처리, 머신러닝 학습              | 웹 크롤링, 대기 중 요청 처리 등             |
| ✅ 구현     | `multiprocessing` 모듈         | `threading` 모듈                  |
