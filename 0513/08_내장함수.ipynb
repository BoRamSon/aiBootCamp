```python
# -------------------------------
# ì ˆëŒ“ê°’ (absolute value)
print('< abs >')  
print(abs(-4))    # 4
print(abs(10))    # 10


# -------------------------------
# all() : ëª¨ë“  ìš”ì†Œê°€ ì°¸ì´ë©´ True
print('< all >') 
print(all([1, 2, 3]))       # True
print(all([1, 2, 3, 0]))    # False (0ì€ False ì·¨ê¸‰)
print(all(['a', 'b', 'c'])) # True
print(all(['a', 'b', '']))  # False (''ëŠ” False)

# all()ì€ ë¹ˆ ë¦¬ìŠ¤íŠ¸ì¼ ê²½ìš° Trueë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤
print(all([]))              # True


# -------------------------------
# any() : í•˜ë‚˜ë¼ë„ ì°¸ì´ë©´ True
print('< any >')
print(any([1, 2, 3]))       # True
print(any([1, 2, 3, 0]))    # True
print(any([0, "", False]))  # False
print(any([]))              # False


# -------------------------------
# ord() / chr()
print('< ord / chr >')
# ord : ë¬¸ìë¥¼ ìœ ë‹ˆì½”ë“œ ì •ìˆ˜ë¡œ
print(ord('A'))  # 65
print(ord('a'))  # 97

# chr : ìœ ë‹ˆì½”ë“œ ì •ìˆ˜ë¥¼ ë¬¸ìë¡œ
print(chr(65))   # 'A'
print(chr(97))   # 'a'


# -------------------------------
# dir() : ê°ì²´ì˜ ì†ì„± ë° ë©”ì„œë“œ ëª©ë¡ì„ ë°˜í™˜
print('< dir >')
print(dir([1, 2, 3]))   # ë¦¬ìŠ¤íŠ¸ê°€ ê°€ì§„ ë©”ì„œë“œë“¤
print(dir(dict()))      # ë”•ì…”ë„ˆë¦¬ê°€ ê°€ì§„ ë©”ì„œë“œë“¤


# -------------------------------
# divmod(a, b) : ëª«ê³¼ ë‚˜ë¨¸ì§€ë¥¼ íŠœí”Œë¡œ ë°˜í™˜
print('< divmod >')
print(divmod(5, 2))   # (2, 1) -> 5ë¥¼ 2ë¡œ ë‚˜ëˆˆ ëª«ê³¼ ë‚˜ë¨¸ì§€
print(divmod(9, 3))   # (3, 0)


# -------------------------------
# enumerate() : ì¸ë±ìŠ¤ì™€ ê°’ì„ í•¨ê»˜ ë°˜í™˜í•˜ëŠ” ë°˜ë³µë¬¸
print('< enumerate >')
for i, c in enumerate("Life is egg"):
    print(i, c)


# -------------------------------
# eval() : ë¬¸ìì—´ì„ íŒŒì´ì¬ ì½”ë“œë¡œ ì‹¤í–‰
print('< eval >')
result = eval('1 + 10 + 3')   # 14
print(result)

# ë³€ìˆ˜ë„ ì“¸ ìˆ˜ ìˆìŒ
x = 5
print(eval('x + 10'))         # 15


# -------------------------------
# filter(í•¨ìˆ˜, iterable) : ì¡°ê±´ì„ ë§Œì¡±í•˜ëŠ” ìš”ì†Œë§Œ í•„í„°ë§
print('< filter >')
a = [3.4, -1, 2, 9, 8, 7, 12, 15, 21]

# ì¼ë°˜ í•¨ìˆ˜ ì‚¬ìš©
def is_positive(x):
    return x > 0

po_list = list(filter(is_positive, a))
print(po_list)  # [3.4, 2, 9, 8, 7, 12, 15, 21]

# lambda (ìµëª… í•¨ìˆ˜) ì‚¬ìš©
po_list2 = list(filter(lambda x: x > 0, a))
print(po_list2)


# -------------------------------
# max() / min() : ìµœëŒ€ê°’ / ìµœì†Œê°’
print('< max / min >')
print(f"ìµœëŒ€ê°’: {max(a)} | ìµœì†Œê°’: {min(a)}")


# -------------------------------
# pow(x, y) : xì˜ yì œê³±ì„ ê³„ì‚° (x ** yì™€ ë™ì¼)
print('< pow >')
print(pow(2, 4))  # 2^4 = 16
print(pow(3, 3))  # 3^3 = 27


# -------------------------------
# round() : ë°˜ì˜¬ë¦¼ (ì†Œìˆ˜ì  ìë¦¬ìˆ˜ ì§€ì • ê°€ëŠ¥)
print('< round >')
print(round(3.14159))      # 3
print(round(3.14159, 2))   # 3.14
print(round(1.5555, 3))    # 1.556

# ì°¸ê³ : íŒŒì´ì¬ì˜ roundëŠ” "ì€í–‰ ë°˜ì˜¬ë¦¼" ë°©ì‹ (0.5ëŠ” ì§ìˆ˜ë¡œ ë°˜ì˜¬ë¦¼)
print(round(0.5))          # 0
print(round(1.5))          # 2
```
