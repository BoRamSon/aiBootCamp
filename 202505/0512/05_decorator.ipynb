````python
def decorator1(func):   # ë§¤ê°œë³€ìˆ˜ëŠ” ì¤‘ì²©í•¨ìˆ˜ ì•ˆì—ì„œ í˜¸ì¶œë  í•¨ìˆ˜
  def wrapper():
    print('í•¨ìˆ˜í˜¸ì¶œ ì „')
    func()
    print('í•¨ìˆ˜í˜¸ì¶œ í›„')
  return wrapper   # ì¤‘ì²©í•¨ìˆ˜ì˜ ì°¸ì¡°ë¥¼ ë°˜í™˜í•´ì•¼ í•œë‹¤.

@decorator1   # ì•„ë˜ hello ì‹¤í–‰í•˜ë©´ decorator1ì´ ë‚©ì¹˜í•œë‹¤. (ë¬´ìŠ¨ ì›ë¦¬ì¸ì§€ ëª¨ë¥´ê² ë‹¤.)
def hello():
  print('hello')

hello()
# decorator1ì—ê²Œ ë‚©ì¹˜ë‹¹í•¨
# -> wrapper -> func()ë¥¼ í†µí•´ í•¨ìˆ˜ í˜¸ì¶œ
````

````python
import time

def time_decorator(callback):
  def innerfunction():
    start = time.time()
    callback()
    end = time.time()
    print(f"{callback.__name__} | ì‹¤í–‰ì‹œê°„: {end-start}ì´ˆ")
  return innerfunction

@time_decorator
def sigma():
  s = 1
  for i in range (1, 10000000):
    s += i
  print("í•©ê³„:", s)

sigma()

````

````python
def mylog(callback):
  def wrapper(*args, **kargs):
    result = callback(*args, **kargs)
    return result
  return wrapper

@mylog
def sigma2(x):
  target = 0
  for i in range(1, x+1):
    target += i
  return target

print(sigma2(10))
````

````python
def mylog(callback):
  def wrapper(*args, **kargs):
    result = callback(*args, **kargs)
    print(f"[LOG] í•¨ìˆ˜ì´ë¦„: {callback.__name__}")  # ğŸ”¥ ì´ë ‡ê²Œ í•´ì„œ callbackìœ¼ë¡œ ë“¤ì–´ì˜¨ í•¨ìˆ˜ì˜ ëª…ì¹­ì„ ì•Œ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
    print(f"[LOG] ì…ë ¥ê°’: args {args} kargs {kargs}")
    print(f"[LOG] ë°˜í™˜ê°’: {result}")
    return result  # ë°˜ë“œì‹œ ë°˜í™˜ì„ í•´ì•¼í•œë‹¤.
  return wrapper

@mylog
def sigma2(x):
  target = 0
  for i in range(1, x+1):
    target += i
  return target

print(sigma2(100))
print(sigma2(10))


# ê³„ì† ì“¸ ìˆ˜ ìˆìŠµë‹ˆë‹¤. 
@mylog
def sub(a, b):
  return a-b    
print(sub(3, 4))

# ê²°ê³¼
# [LOG] í•¨ìˆ˜ì´ë¦„: sub
# [LOG] ì…ë ¥ê°’: args (3, 4) kargs {}
# [LOG] ë°˜í™˜ê°’: -1
# -1

# = ì´ë ‡ê²Œ ë°˜ë³µì ì¸ ì‘ì—…ë“¤ì— ê³µí†µì ìœ¼ë¡œ ë“¤ì–´ê°€ëŠ” ê²ƒì— decoratorë¥¼ ì”Œì›Œë†“ìœ¼ë©´ ì•Œì•„ì„œ ê°€ë¡œì±„ê°€ì„œ ê³ ì •ì ì¸ ì •ë³´ë“¤ì„ ì¶œë ¥í•´ì£¼ëŠ” ê²ƒì„ ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
````
