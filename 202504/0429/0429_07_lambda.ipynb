## ğŸŸ¢ í•¨ìˆ˜ì˜ ë§¤ê°œë³€ìˆ˜ë¡œ í•¨ìˆ˜ë¥¼ ì¤„ ìˆ˜ ìˆìŒ

````python
# 1. ê³ ì°¨ í•¨ìˆ˜: í•¨ìˆ˜ë¥¼ ì¸ìë¡œ ë°›ì„ ìˆ˜ ìˆëŠ” í•¨ìˆ˜ ì •ì˜
def myfunc(x, y, callback):  # ì„¸ ë²ˆì§¸ ë§¤ê°œë³€ìˆ˜ëŠ” 'í•¨ìˆ˜(ì½œë°± í•¨ìˆ˜)'ë¥¼ ì „ë‹¬ë°›ìŒ
    result = callback(x, y)  # ì „ë‹¬ë°›ì€ í•¨ìˆ˜ë¥¼ ì‹¤í–‰í•´ì„œ ê²°ê³¼ë¥¼ resultì— ì €ì¥
    print(x, y, result)      # x, y, ê³„ì‚° ê²°ê³¼ ì¶œë ¥

# 2. ì¼ë°˜ í•¨ìˆ˜ ì •ì˜ (ë‘ ìˆ˜ë¥¼ ë”í•¨)
def add(x, y):
    return x + y

# 3. í•¨ìˆ˜ í˜¸ì¶œ - add í•¨ìˆ˜ë¥¼ ì½œë°±ìœ¼ë¡œ ì „ë‹¬
myfunc(4, 5, add)  # 4 + 5 â†’ 9 ì¶œë ¥

# 4. ğŸ”¥ğŸ”¥ğŸ”¥ğŸ”¥ğŸ”¥ ëŒë‹¤ í•¨ìˆ˜ë¥¼ ì§ì ‘ ì „ë‹¬ (í•¨ìˆ˜ ì •ì˜ ì—†ì´ ì¦‰ì„ì—ì„œ ëº„ì…ˆ ê¸°ëŠ¥ ì •ì˜)
myfunc(4, 5, lambda x, y: x - y)  # 4 - 5 â†’ -1 ì¶œë ¥   # ê²°ê³¼ : 4 5 -1
````

````python
# ëŒë‹¤ í•¨ìˆ˜ ì—¬ëŸ¬ ê°œë¥¼ ë¦¬ìŠ¤íŠ¸ë¡œ ì €ì¥
fuc_list = [
    lambda x, y: x - y,  # 0ë²ˆ ì¸ë±ìŠ¤: ëº„ì…ˆ í•¨ìˆ˜
    lambda x, y: x + y,  # 1ë²ˆ ì¸ë±ìŠ¤: ë§ì…ˆ í•¨ìˆ˜
    lambda x, y: x * y,  # 2ë²ˆ ì¸ë±ìŠ¤: ê³±ì…ˆ í•¨ìˆ˜
    lambda x, y: x / y   # 3ë²ˆ ì¸ë±ìŠ¤: ë‚˜ëˆ—ì…ˆ í•¨ìˆ˜
]

# ë¦¬ìŠ¤íŠ¸ì— ì €ì¥ëœ í•¨ìˆ˜ë¥¼ ë°˜ë³µ ì‹¤í–‰
for f in fuc_list:       # ë¦¬ìŠ¤íŠ¸ì˜ ê° í•¨ìˆ˜(f)ë¥¼ êº¼ë‚´ì„œ
    print(f(9, 7))       # f(9, 7)ì„ ì‹¤í–‰ â†’ ê°ê° ëº„ì…ˆ, ë§ì…ˆ, ê³±ì…ˆ, ë‚˜ëˆ—ì…ˆ ê²°ê³¼ ì¶œë ¥
````
