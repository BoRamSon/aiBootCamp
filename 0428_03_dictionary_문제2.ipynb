# 🟩 리스트, 딕셔너리 문제 풀기

- 반에 5명이 존재한다. 이 인원을 대상으로 

- <입력>
  - 이름 
  - 국어
  - 영어
  - 수학

- <계산>
  - 1. 총점
  - 2. 평균
  - 3. 평균 결과에 대해서 
    - 수(90)
    - 우(80)
    - 미(70)
    - 양(60)
    - 가(60미만)  

- <출력>
  - 1. 총점
  - 2. 평균
  - 3. 평균 결과에 대해서 
    - 수(90)
    - 우(80)
    - 미(70)
    - 양(60)
    - 가(60미만)





---
## 🟢 직접 리스트만 사용해서 풀어보기

```python
# ==============================================
# 입력
name_list = []
korean_lang_sub_list = []
english_lang_sub_list = []
math_sub_list = []

for i in range(0, 1):
  name = input('이름을 입력하시오 : ')
  korean_lang_sub = int(input('국어 점수를 입력하시오 : '))
  english_lang_sub = int(input('영어 점수를 입력하시오 : '))
  math_sub = int(input('수학 점수를 입력하시오 : '))

  name_list.append(name)
  korean_lang_sub_list.append(korean_lang_sub)
  english_lang_sub_list.append(english_lang_sub)
  math_sub_list.append(math_sub)


# ==============================================
# 계산
grade_sum_result = []
grade_average_result = []
grade_average_korean_result = []

for i in range(0, 1):
  total_grade = korean_lang_sub_list[i] + english_lang_sub_list [i] + math_sub_list[i]
  grade_sum_result.append(total_grade)
  total_grade_average = total_grade / 3
  grade_average_result.append(total_grade_average)
  if total_grade_average >= 90:
    grade_average_korean_result.append('(수)')
  elif total_grade_average >= 80:
    grade_average_korean_result.append('(우)')
  elif total_grade_average >= 70:
    grade_average_korean_result.append('(미)')
  elif total_grade_average >= 60:
    grade_average_korean_result.append('(양)')
  else:
    grade_average_korean_result.append('(가)')


# ==============================================
# 출력
for i in range(0, 1):
  print(f"{name_list[i]}님의 총점은 {grade_sum_result[i]}이고, 평균은 {grade_average_result[i]}이며, 한글점수는 {grade_average_korean_result[i]} 입니다.")


```

## 🟢 같이 풀어보기 (리스트 형태)

```python
# ==============================================
# 입력
name_list = []
korean_lang_sub_list = []
english_lang_sub_list = []
math_sub_list = []

for i in range(0, 1):
  name = input('이름을 입력하시오 : ')
  korean_lang_sub = int(input('국어 점수를 입력하시오 : '))
  english_lang_sub = int(input('영어 점수를 입력하시오 : '))
  math_sub = int(input('수학 점수를 입력하시오 : '))

  name_list.append(name)
  korean_lang_sub_list.append(korean_lang_sub)
  english_lang_sub_list.append(english_lang_sub)
  math_sub_list.append(math_sub)


# ==============================================
# 계산
grade_sum_result = []
grade_average_result = []
grade_average_korean_result = []

for i in range(0, len(name_list)):
  total_grade = korean_lang_sub_list[i] + english_lang_sub_list [i] + math_sub_list[i]
  grade_sum_result.append(total_grade)

  total_grade_average = total_grade / 3
  grade_average_result.append(total_grade_average)

print(grade_average_result)
  
for i in range(0, len(name_list)):  
  if grade_average_result[i] >= 90:
    grade_average_korean_result.append('(수)')
  elif grade_average_result[i] >= 80:
    grade_average_korean_result.append('(우)')
  elif grade_average_result[i] >= 70:
    grade_average_korean_result.append('(미)')
  elif grade_average_result[i] >= 60:
    grade_average_korean_result.append('(양)')
  else:
    grade_average_korean_result.append('(가)')


# ==============================================
# 출력
for i in range(0, len(name_list)):
  
  # print가 너무 길고 복잡하면 이렇게 나눌 수 있스니다.
  # 

  print(f"{name_list[i]}님,", end="\t")
  print(f"{korean_lang_sub_list}", end="\t")
  print(f"{english_lang_sub_list}", end="\t")
  print(f"{math_sub_list}", end="\t")
  print(f"총점은 {grade_sum_result[i]}이고,", end="\t") 
  print(f"평균은 {grade_average_result[i]}이며,", end="\t")
  print(f"한글점수는 {grade_average_korean_result[i]} 입니다.")

```

## 🟢 dic으로 푼 문제 같이 풀어보기

```python

# ------------------------------------------------------
# 입력
score_list = []

for i in range(0, 4):
  name = input('이름을 입력하시오 : ')
  kor = int(input('국어 점수를 입력하시오 : '))
  eng = int(input('영어 점수를 입력하시오 : '))
  mat = int(input('수학 점수를 입력하시오 : '))

  score={} 
  score["name"]= name  # 🔥 이게 키와 값인 것입니다. 이게 단순 추가인 것입니다!!!
  score["kor"] = kor 
  score["eng"] = eng 
  score["mat"] = mat

  score_list.append( score ) 


# ------------------------------------------------------
# 계산
for i in range(0, len(score_list)):
  score = score_list[i] # 🔥 저장되었던 dict타입 객체를 가져오는 것
                      # 🔥 이것은 리스트 내 딕타입 하나만 가져와서 계산을 하고 있는 것이다. 나는 완전 새롭게 
                      # 🔥 내가 왜 이런 생각하지 못했을까....
                      # ❓❓❓❓❓ 아닌데... 내가 위에서 한거랑 똑같은데

  score['total'] = score['kor'] + score['eng'] + score['mat']

  score['avg'] = score['total'] // 3 # 몫만 구하기

for i in range(0, len(score_list)):
    score = score_list[i] #저장되었던 dict타입 객체를 가져와서 
    if score["avg"]>=90:
        score["grade"] = "수"
    elif score["avg"]>=80:   
        score["grade"] = "우"     
    elif score["avg"]>=70:   
        score["grade"] = "미"     
    elif score["avg"]>=60:   
        score["grade"] = "양" 
    else:
        score["grade"] = "가"    


# ------------------------------------------------------
# 출력
for score in score_list:
  print(f"{score['name']}", end='\t')
  print(f"{score['kor']}", end='\t')
  print(f"{score['eng']}", end='\t')
  print(f"{score['mat']}", end='\t')
  print(f"{score['total']}", end='\t')
  print(f"{score['avg']}", end='\t')
  print(f"{score['grade']}")
```
