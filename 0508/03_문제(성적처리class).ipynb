```python
import re

class InputValidator:  
  def __init__(self):
    pass
    # self.user_input = user_input

  def validate_string(self, input):
    if self.is_empty(input):
      raise ValueError('ë¹ˆ ê°’ì…ë‹ˆë‹¤. ë‹¤ì‹œ ì…ë ¥í•´ì£¼ì„¸ìš”.')
    if self.is_numeric(input):
      raise ValueError('ìˆ«ìê°€ ì…ë ¥ë  ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ë‹¤ì‹œ ì…ë ¥í•´ì£¼ì„¸ìš”.')
    if self.is_korean_string(input):
      raise ValueError('í•œê¸€ë§Œ ì…ë ¥í•  ìˆ˜ ìˆì–´ìš”. ë‹¤ì‹œ ì…ë ¥í•´ì£¼ì„¸ìš”.')
    
  def is_numeric(self, input):
    return input.isnumeric()
  
  def is_korean_string(self, input):
    return not bool(re.fullmatch(r"[ê°€-í£]+", input))

  def validate_number(self, input):
    if self.is_empty(input):
      raise ValueError('ë¹ˆ ê°’ì…ë‹ˆë‹¤. ë‹¤ì‹œ ì…ë ¥í•´ì£¼ì„¸ìš”.')
    # if self.is_number_type(input):
    #   raise ValueError('ìˆ«ìë§Œ ì…ë ¥í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ë‹¤ì‹œ ì…ë ¥í•´ì£¼ì„¸ìš”.')
    if self.is_negative(input):
      raise ValueError('ìŒìˆ˜ëŠ” í—ˆìš©ë˜ì§€ ì•ŠìŠµë‹ˆë‹¤. ë‹¤ì‹œ ì…ë ¥í•´ì£¼ì„¸ìš”.') 

  def is_empty(self, input):
    return input == ''

  # def is_number_type(self, input):
  #   return isinstance(input, (int))

  def is_negative(self, input):
    return int(input) < 0
  
  def input_validation(self, select_type, dict):
    for key, message in dict.items():
      for i in range(1,11):
        validate_target = input(message)
        try:
          if select_type == 'string':
            self.validate_string(validate_target)
          elif select_type == 'number':
            self.validate_number(validate_target)
        except ValueError as e:
          print('ERROR:', e)
          if i == 10: return print('ì¢…ë£Œë˜ì—ˆìŠµë‹ˆë‹¤. ë‹¤ì‹œ ì‹œë„í•˜ì„¸ìš”.')
        else:
          return {key : validate_target}


class InputDataAndValidate:
  def __init__(self):
    self.input_validator = InputValidator()
    self.INPUT_MESSAGES_STRING = {'name': 'í•™ìƒì˜ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”. :'}
    self.INPUT_MESSAGES_NUMBER = {'kor_score': 'êµ­ì–´ ì„±ì ì„ ì…ë ¥í•˜ì„¸ìš”. :', 
                                  'eng_score': 'ì˜ì–´ ì„±ì ì„ ì…ë ¥í•˜ì„¸ìš”. :', 
                                  'mth_score': 'ìˆ˜í•™ ì„±ì ì„ ì…ë ¥í•˜ì„¸ìš”. :' 
    }

  def input_string(self, string_dict):
    return self.input_validator.input_validation('string', string_dict)

  # def input_string(self, string_dict):
  #   name = {}
  #   for key, message in string_dict.items():

  #     # name = self.vaildation(message)

  #     for i in range(1,11):
  #       name[key] = input(message)
  #       try:
  #         validator = InputValidator(name[key])
  #         validator.validate_string()
  #       except ValueError as e:
  #         print('ERROR:', e)
  #         if i == 10: return print('ì¢…ë£Œë˜ì—ˆìŠµë‹ˆë‹¤. ë‹¤ì‹œ ì‹œë„í•˜ì„¸ìš”.')
  #       else:
  #         break

  #     print(f"ì‘ë™ì¤‘: \'{key}\': \'{name[key]}\'")

  #   return name
  
  def input_number(self, number_dict):
    score = {}
    for key, message in number_dict.items():
      for i in range(1,11):
        score[key] = int(input(message))
        try:
          validator = InputValidator()
          validator.validate_number(score[key])
        except ValueError as e:
          print('ERROR:', e)
          if i == 10: return print('ì¢…ë£Œë˜ì—ˆìŠµë‹ˆë‹¤. ë‹¤ì‹œ ì‹œë„í•˜ì„¸ìš”.')
        else:
          break
      print(f"ì‘ë™ì¤‘: \'{key}\': \'{score[key]}\'")
    return score

  def input_result(self):
    a = self.input_string(self.INPUT_MESSAGES_STRING)
    print(a)
    b = self.input_number(self.INPUT_MESSAGES_NUMBER)
    print(b)
    # ğŸŸ¡ ë”•ì…”ë„ˆë¦¬ë¼ë¦¬ í•©ì¹˜ëŠ” ë°©ë²• 1
    # a.update(b)
    # ğŸŸ¡ ë”•ì…”ë„ˆë¦¬ë¼ë¦¬ í•©ì¹˜ëŠ” ë°©ë²• 2 (3.9 ì´ìƒë¶€í„°)
    student = a | b
    print(f"ì™„ì„±ëœ dict: {student} ì…ë‹ˆë‹¤.")
    return student  # dict íƒ€ì… ë°˜í™˜í•´ì£¼ëŠ” ì¹œêµ¬


class CalculGrade():
  def __init__(self):
    pass

  def cal_score(self, student) :
    student["score_sum"] = student["kor_score"] + student["eng_score"] + student["mth_score"]
    student["score_aver"] = student["score_sum"] // 3
    return {student["score_sum"], student["score_aver"]}
  
  def cal_grade(self, student) :
    if student["score_aver"] >= 80 :
      student["grade"] = "A"
    elif student["score_aver"] >= 70 :
      student["grade"] = "B"
    elif student["score_aver"] >= 60 :
      student["grade"] = "C"
    else :
      student["grade"] = "D"
    return {student["grade"]}  

  def cal_main(self, calcul_dict_list) :
    for i in calcul_dict_list :
        self.cal_score(i)
        self.cal_grade(i)


class OutputData():
  def __init__(self):
    pass

  def output_menu(self):
    print("[1] í•™ìƒ ì…ë ¥")
    print("[2] ì„±ì  ê³„ì‚°")
    print("[3] ê¸°ë¡ ì¶œë ¥")
    print("[0] í”„ë¡œê·¸ë¨ ì¢…ë£Œ")

  def output_student(self, output_student_dict_list) :
    for i in output_student_dict_list :
      print(f"ì´ë¦„ : {i['name']}",end="\t")
      print(f"êµ­ì–´ ì„±ì  : {i['kor_score']}",end="\t")
      print(f"ì˜ì–´ ì„±ì  : {i['eng_score']}",end="\t")
      print(f"ìˆ˜í•™ ì„±ì  : {i['mth_score']}",end="\t")
      print(f"ì´ ì  : {i['score_sum']}",end="\t")
      print(f"í‰ ê·  : {i['score_aver']:.0f}",end="\t")
      print(f"ë“± ê¸‰ : {i['grade']}")


class Controller():
  def __init__(self):
    self.input_student = InputDataAndValidate()
    self.calcul_student = CalculGrade()
    self.output_student = OutputData()
    self.student_grade_list = []

  def start(self):
    while (True) :
      self.output_student.output_menu()
      sel = input("ë©”ë‰´ ì„ íƒ : ")
      if sel == "1" :
          self.student_grade_list.append(self.input_student.input_result())
      elif sel == "2" :
          self.calcul_student.cal_main(self.student_grade_list)
      elif sel == "3" :
          self.output_student.output_student(self.student_grade_list)
      elif sel == "0" :
          print('ì •ìƒì ìœ¼ë¡œ ì¢…ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.')    
          return
      else :
          print('ì˜ëª» ì…ë ¥í•˜ì…¨ìŠµë‹ˆë‹¤.')


controller = Controller()
controller.start()
```
