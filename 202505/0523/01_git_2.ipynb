# 🟩 Git & Github

## 🟢 Git - HomePC, OfficePC 실습해보기

### 🟡 원격저장소 연결하기
- 나의 홈디렉토리에 특정 폴더(practice_git) 생성
  - cd practice_git

- home_pc 디렉터리에 원격저장소 복제하기
  - git clone 원격_저장소_주소 home_pc
    - (home_pc는 이 폴더를 만들어서 원격저장소를 넣어달라는 의미)

- office_pc 디렉터리에 원격저장소 복제하기
  - git clone 원격_저장소_주소 office_pc
  
- 원격저장소가 복제 되었는지 확인
  - cd home_pc

- 로그 확인하기
  - git log

- 지역저장소가 원격저장소와 연결되었는지 확인
  - git remote -v

### 🟡 먼저 home_pc폴더에서 작업하기
- home_pc 디렉터리로 이동
  - cd home_pc
- 새 파일 만들고, 내용 입력
  - vim firstHome.txt
  - 내용 입력 ex. add
- git add firstHome.txt
- git commit -m "add firstHome"
- git push
- 깃허브 웹을 확인하여 

### 🟡 office_pc폴더에서 작업하기
- office_pc 디렉터리로 이동
  - cd office_pc
- git pull
- firstHome.txt 파일 수정
  - vim firstHome.txt
- git add firstHome.txt
- git commit -m "add 4"

## 🟢 Fork
- 다른 사람의 저장소(repository)를 내 GitHub 계정으로 복사하는 것

### 🟡 나의 Local로 복사해오기
- 나의 Local Project Folder(ex. practice_git)
- git clone 복사한_저장소_주소 game
  - cd game
  - ls -la
- 소스코드 수정하여, add, commit, push 등을 통해 나의 Fork된 Repository를 자유롭게 사용할 수 있습니다.


## 🟢 Pull Request
- 그런데 수정을 하고 Fork했던 Repository의 원본소유자에서 반영시켜주고 싶다고 가정해보자. 이럴 때 원본소유자에게 이렇게 반영하면 좋을 것 같다는 의견을 남기는 것이 바로 pull request 입니다.

- Fork된 Repository  >  Pull Requests  >  New Pull Request  >  제목, 설명 등 작성  >  Create Pull Request

- 원본소유자는 해당 PR을 읽고, 소스코드를 확인하여 최종적으로 Merge pull request를 클릭하여 해당 코드를 합칩니다.
