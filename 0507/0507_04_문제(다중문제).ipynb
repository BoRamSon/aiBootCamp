````python
def duplicate_remove(aList):
    temp = []
    for i in aList:
        if i not in temp:  # tempì•ˆì— ì¡´ì¬ì•ˆí•  ë•Œ
            temp.append(i)
    return temp     

a = [4,3,5,8,1,2,56,4,1,2,8]
b = duplicate_remove(a)
print(b)
````

### ğŸŸ¡ ê°™ì´ í’€ì–´ë³´ê¸°

````python
# ordë¥¼ í†µí•´ì„œ ìœ ë‹ˆì½”ë“œ ì‚¬ìš©í•˜ì—¬ í’€ì–´ë³´ê¸°

def myint(s):
  sum = 0    # 123  1 2  =>  1*10 + 2 => 123 => 12*10+3
  for c in s:
    if ord(c) < ord('0') or ord(c) > ord('9'):  # ì¦‰, ìˆ«ì '0'ë¶€í„° '9'ê¹Œì§€ì˜ ìœ ë‹ˆì½”ë“œ(ASCII ì½”ë“œ)ëŠ” 48~57ì…ë‹ˆë‹¤.
      return -1
    sum = sum * 10 + ord(c)-ord('0')   # ë¬¸ì1 -> ìˆ«ì 1ë¡œ ë¹ êµ¬ë ¤ë¨¼ '1'-'0'
  return sum

print(myint('123') + myint('345'))
````

````python
def reverse(s):
  result="";
  for i in range(len(s)-1, -1, -1):
      result += s[i] 
  return result 

print( reverse("korea"))
````
