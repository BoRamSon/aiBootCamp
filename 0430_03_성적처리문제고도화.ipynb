```python
import random
# ì™¸ë¶€ ëª¨ë“ˆ(íŒŒì´ì¬ì˜ ê²½ìš° íŒŒì¼) ì„ ì´ ë©”ëª¨ë¦¬ë¡œ ê°€ì ¸ì™€ì„œ ê²°í•©(ì‘ì—…) í•˜ë¼ëŠ” ëª…ë ¹ì–´
# random ì´ë¼ëŠ” í•¨ìˆ˜ëŠ” ì •ìˆ˜ ë°œìƒ í•¨ìˆ˜

score_list = []


# ==========================================================


def init() :     # ë§¨ ì²˜ìŒì— ê¸°ë³¸ ë°ì´í„°ë¥¼ ì…ë ¥í•´ë†“ê³  ì‹œì‘í•˜ê¸° ìœ„í•œ í•¨ìˆ˜
    names_list = ["í™ê¸¸ë™" , "ê°•ê°ì°¬" , "ê³½ì¬ìš°" , "í™ê²½ë˜" , "ì„œí¬" ,"ìœ¤ê´€" , "ê¹€ì—°ì•„" , "ì•ˆì„¸ì˜" , "ì¡°ìŠ¹ì—°" , "ì¥ê¸¸ì‚°"]
    for i in range(0,10) :
        score_list.append({"name" : names_list[i] , "kor" : random.randint(20,100) , "eng" : random.randint(20,100) , "mth" : random.randint(20,100)})
    for s in score_list :
        s["sum"] = getSum(s)
        s["avg"] = getAvg(s)
        s["grade"] = getGrade(s)

def append() :
    s = {}
    
    s["name"]  = input("ì´ë¦„ ì…ë ¥ : ")
    s["kor"]   = getScore("êµ­ì–´")
    s["eng"]   = getScore("ì˜ì–´")
    s["mth"]   = getScore("ìˆ˜í•™")
    s["sum"]   = getSum(s)
    s["avg"]   = getAvg(s)
    s["grade"] = getGrade(s)
    
    score_list.append(s)



def getNumber(subject) :
    number = input(f"{subject} : ")
    while isDigit(number) == False :
        print("ìˆ«ìë§Œ ì…ë ¥í•´ì£¼ì„¸ìš”.")
        number = input("ì •ìˆ˜ë¥¼ ì…ë ¥í•˜ì„¸ìš” : ")
    return int(number)

# input ìœ¼ë¡œ ë°›ëŠ” ëª¨ë“  ë°ì´í„°ëŠ” string íƒ€ì…
# ord í•¨ìˆ˜ë¥¼ í†µí•´ì„œ ìˆ«ìì¸ì§€ ì•„ë‹Œì§€ íŒë‹¨ ê°€ëŠ¥
# ë¬¸ìì—´ì„ ë°›ì•„ì„œ í•œê¸€ìì”© "0" ê³¼ "9" ì‚¬ì´ì— ìˆëŠ” ì§€ í™•ì¸
# ê¸€ì ì¤‘ì— í•˜ë‚˜ë¼ë„ 0 - 9 ì‚¬ì´ì— ì¡´ì¬í•˜ì§€ ì•Šìœ¼ë©´ ì—ëŸ¬(ìˆ«ìê°€ ì•„ë‹Œ ê²ƒìœ¼ë¡œ íŒë‹¨)
def isDigit(number) :
    for i in range(0,len(number)) :
        if ord(number[i]) < ord("0") or ord(number[i]) > ord("9") :
            return False
    
    return True



# ==========================================================


# ì…ë ¥í•œ ì ìˆ˜ê°€ 0 ~ 100 ì‚¬ì´ì— ìˆëŠ” ìˆ«ìì¸ì§€ í™•ì¸
def getScore(subject = "êµ­ì–´" , limit = 100) :
  n = getNumber(subject)
  while n < 0 or n > limit :
    print(f"0 ~ {limit} ì‚¬ì´ì˜ ê°’ì„ ì…ë ¥í•˜ì„¸ìš”")
    n = getNumber(subject)
  return n

# êµ­ì–´ , ì˜ì–´ , ìˆ˜í•™ ì„±ì  í•©ì‚°
def getSum(s) :
  return s["kor"] + s["eng"] + s["mth"]

# í•©ì‚° ì„±ì ì˜ í‰ê·  ê³„ì‚°
def getAvg(s) :
  return s["sum"] // 3

# ì„±ì  í‰ê· ì— ë”°ë¥¸ ë“±ê¸‰ ì„¤ì •
def getGrade(s) :
  if   s["avg"] >= 80 :
      return "ìˆ˜"
  elif s["avg"] >= 70 :
      return "ìš°"
  elif s["avg"] >= 60 :
      return "ë¯¸"
  elif s["avg"] >= 50 :
      return "ì–‘"
  else :
      return "ê°€"



# ==========================================================



# filter ì—ì„œ ë‘ë²ˆì§¸ íŒŒë¼ë¯¸í„° iterable (ë°˜ë³µì ì¸) ë°ì´í„°ë¥¼ ì „ë‹¬ - list ë¥˜
# filter ì•ˆì—ì„œ list ì•ˆì—ì„œ í•œ ê°œì”© ê°ì²´ë¥¼ ê°€ì§€ê³  ì˜´
# ë°˜í™˜ê°’ì´ list íƒ€ì…ì´ê³  list ì•ˆì— dict íƒ€ì…ì´ ìˆìŒ
# nameSearch ì—ì„œë„ ouput() í•¨ìˆ˜ë¥¼ ì¶œë ¥í•˜ê³  ì‹¶ìŒ
# í•¨ìˆ˜ ì•ˆì˜ ë§¤ê°œë³€ìˆ˜ì™€ ì™¸ë¶€ ì „ì—­ë³€ìˆ˜ì˜ ì´ë¦„ì´ ë™ì¼í•˜ë©´ ì™¸ë¶€ì˜ ì „ì—­ ë³€ìˆ˜ë³´ë‹¤ í•¨ìˆ˜ ì•ˆì˜ ë§¤ê°œë³€ìˆ˜ë¥¼ ìš°ì„ ì‹œí•¨
def nameSearch() :
  name = input("ê²€ìƒ‰í•˜ë ¤ëŠ” ì´ë¦„ì„ ì…ë ¥í•´ì£¼ì„¸ìš” : ")
  result_list = list(filter(lambda x : x["name"] == name , score_list))
  
  output(result_list)
    
# ì´ë¦„ìœ¼ë¡œ ê²€ìƒ‰í•˜ì—¬ ì‚­ì œ ê¸°ëŠ¥ ì¶”ê°€
def nameDel() :
  name = input("ì‚­ì œí•˜ë ¤ëŠ” ì´ë¦„ì€ ì…ë ¥í•´ì£¼ì„¸ìš” : ")
  del_name = {}
  for i in filter(lambda x : x["name"] == name , score_list) :
    del_name = i
  score_list.remove(del_name)

# ì´ë¦„ìœ¼ë¡œ ê²€ìƒ‰ í›„ ìˆ˜ì • ê¸°ëŠ¥ ì¶”ê°€
def nameModify() :
  name = input("ìˆ˜ì •í•˜ë ¤ëŠ” ì •ë³´ì˜ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš” : ")
  result_list = list(filter(lambda x : x["name"] == name , score_list))
  if len(result_list) == 0 :
    print(f"{name} ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
    return
  
  # ë‘ëª… ì´ìƒ ìˆì„ ë•Œì˜ ì²˜ë¦¬ ë°©ë²•ì€ ê³¼ì œ
  student = result_list[0]
  student["name"]  = input("ìˆ˜ì •í•  ì´ë¦„ì„ ì…ë ¥í•´ì£¼ì„¸ìš” : ")
  student["kor"]   = getScore("êµ­ì–´")
  student["eng"]   = getScore("ì˜ì–´")
  student["mth"]   = getScore("ìˆ˜í•™")
  student["sum"]   = getSum(student)
  student["avg"]   = getAvg(student)
  student["grade"] = getGrade(student)
  print("ì…ë ¥í•œ ì •ë³´ë¡œ ìˆ˜ì •ë˜ì—ˆìŠµë‹ˆë‹¤.")



# ==========================================================



def menuDisplay() :
  print("[1] ì¶”ê°€")
  print("[2] ì¶œë ¥")
  print("[3] ê²€ìƒ‰")
  print("[4] ì‚­ì œ")
  print("[5] ìˆ˜ì •")
  print("[0] ì¢…ë£Œ")


def output(score_list) :
  for s in score_list :   # score_list ë¡œë¶€í„° í•˜ë‚˜ì”© s ë¼ëŠ” ë³€ìˆ˜ì— ì „ë‹¬
    print(f"ì´ë¦„ : {s["name"]}",end="\t")
    print(f"êµ­ì–´ ì„±ì  : {s["kor"]}",end="\t")
    print(f"ì˜ì–´ ì„±ì  : {s["eng"]}",end="\t")
    print(f"ìˆ˜í•™ ì„±ì  : {s["mth"]}",end="\t")
    print(f"ì´ ì  : {s["sum"]}",end="\t")
    print(f"í‰ ê·  : {s["avg"]:.0f}",end="\t")
    print(f"ë“± ê¸‰ : {s["grade"]}")



# ==========================================================


        
def start() :
  while True :
    menuDisplay()
    sel = input("ë©”ë‰´ ì„ íƒ : ")
    if sel   == "1" :
        append()
    elif sel == "2" :
        output(score_list)
    elif sel == "3" :
        nameSearch()  # ğŸ”¥ íŠ¹ì • ì´ë¦„ì— ëŒ€í•œ ì •ë³´ë¥¼ ê°€ì ¸ì˜¤ëŠ” ê¸°ëŠ¥
    elif sel == "4" :
        nameDel()     # ğŸ”¥ íŠ¹ì • ì´ë¦„ì„ ë¦¬ìŠ¤íŠ¸ì—ì„œ ì‚­ì œí•˜ëŠ” ê¸°ëŠ¥
    elif sel == "5" :
        nameModify()  # ğŸ”¥ íŠ¹ì • ì´ë¦„ì˜ ë¦¬ìŠ¤íŠ¸ ë‚´ë¶€ ì •ë³´ë¥¼ ìˆ˜ì •í•˜ëŠ” ê¸°ëŠ¥
    elif sel == "0" :
        print("í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
        return
    else :
        print("ë©”ë‰´ ì…ë ¥ì„ ì˜ëª»í•˜ì…¨ìŠµë‹ˆë‹¤.")


init()         
start()
```
